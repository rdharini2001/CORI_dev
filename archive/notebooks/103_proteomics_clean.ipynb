# Proteomics comparison — CORI versus M-MACE

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(PROJECT_ROOT / 'src'))

from cori_analysis.proteomics import ProteomicsConfig, run_proteomics
```

```python
config = ProteomicsConfig(
    source_file=PROJECT_ROOT / 'data' / 'source_population_with_retinal_scores.csv',
    clinical_file=PROJECT_ROOT / 'data' / 'final_df_HTN_DB_Status.csv',
    proteomics_file=PROJECT_ROOT / 'data' / 'proteomics_50k_instance_0_sdf.csv',
    protein_columns_file=PROJECT_ROOT / 'data' / 'alz_proteomics_columns.txt',
    output_dir=PROJECT_ROOT / 'figures' / 'proteomics',
)
results = run_proteomics(config)
```

```python
top_cori = results['CORI'].query('pval_fdr < 0.01').sort_values('log2FC').head(10)
top_mmace = results['MMACE'].query('pval_fdr < 0.01').sort_values('log2FC').head(10)
display(top_cori, top_mmace)
```
