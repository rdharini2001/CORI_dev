```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(PROJECT_ROOT / 'src'))

from cori_analysis.mediation import MediationConfig, run_mediation_pipeline
```

## Primary mediation model

```python
primary = MediationConfig(
    source_file=PROJECT_ROOT / 'data' / 'source_population_with_retinal_scores.csv',
    clinical_file=PROJECT_ROOT / 'data' / 'final_df_HTN_DB_Status.csv',
    output_dir=PROJECT_ROOT / 'figures' / 'mediation_results',
    covariate_formula='age + female + height + C(center) + HTN + Diabetes',
    n_rep=500,
    seed=42,
)
run_mediation_pipeline(primary)
```

## Cancer-status-adjusted mediator sensitivity analysis

```python
cancer_adjusted = MediationConfig(
    source_file=primary.source_file,
    clinical_file=primary.clinical_file,
    output_dir=PROJECT_ROOT / 'figures' / 'mediation_results_cancerAdjusted',
    covariate_formula='age + female + height + C(center) + HTN + Diabetes + A_cancer',
    n_rep=50,
    seed=42,
)
run_mediation_pipeline(cancer_adjusted)
```
